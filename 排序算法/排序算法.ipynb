```python
#冒泡排序
def bubble_sort(arr):
    n=len(arr)
    for i in range(n):
        flag=False
        for j in range(n-i-1):#每次把最大的值放到未排序的最后面
            if arr[j]>arr[j+1]:
                arr[j],arr[j+1]=arr[j+1],arr[j]
                flag=True
        if not flag:
            break
    return arr
```

```python
#选择排序
def selection_sort(arr):
    n=len(arr)
    for i in range(n):
        min_idx=i #每次找出未排序的最小的数放到排好序的部分的最后面
        for j in range(i+1,n):
            if arr[j]<arr[min_idx]:
                min_idx=j
        arr[min_idx],arr[i]=arr[i],arr[min_idx]
    return arr
```

```python
#插入排序
def insertion_sort(arr):
    n=len(arr)
    for i in range(1,n):
        j=i-1
        key=arr[i] #当前要插入的值
        while j>=0 and key<arr[j]: #找到第一个比当前插入的值小的值
            arr[j+1]=arr[j] 
            j-=1
        arr[j+1]=key
    return arr
```

```python
#归并排序
def merge_sort(arr):
    #拆到最细
    if len(arr)<=1:
        return arr
    #先拆成差不多相等的两部分
    mid=len(arr)//2
    left=merge_sort(arr[:mid])
    right=merge_sort(arr[mid:])
    #再把排好序的两部分合并
    return merge(left,right)

def merge(left,right):
    res=[]
    m,n=len(left),len(right)
    i=j=0
    while i<m and j<n:
        if left[i]<=right[j]:
            res.append(left[i])
            i+=1
        else:
            res.append(right[j])
            j+=1
    if i==m:
        res.extend(right[j:])
    if j==n:
        res.extend(left[i:])
    return res
```

```python
#快速排序
def quick_sort(arr): #用第一个值作为基准值
    if len(arr)<=1:
        return arr
    pivot=arr[0]
    left=[x for x in arr[1:] if x<=pivot]
    right=[x for x in arr[1:] if x>pivot]
    return quick_sort(left)+[pivot]+quick_sort(right)
```

```python
#计数排序
def counting_sort(arr):
    if not arr:
        return arr
    min_num=min(arr)
    max_num=max(arr)
    cnt=[0]*(max_num-min_num+1)
    for num in arr:
        cnt[num-min_num]+=1
    res=[]
    for i,num in enumerate(cnt):
        res+=[min_num+i]*num
    return res
```

```python
arr=[64, 34, 25, 12, 22, 11, 90]
print("冒泡排序",bubble_sort(arr))
print("选择排序",selection_sort(arr))
print("插入排序",insertion_sort(arr))
print("归并排序",merge_sort(arr))
print("快速排序",quick_sort(arr))
print("计数排序",counting_sort(arr))
```

```python
import random

def findKthLargest_pythonic(nums, k):
    if not nums: return None
    
    # 随机选择一个基准值
    pivot = random.choice(nums)
    
    # 将数组分成三部分：大于基准、等于基准、小于基准
    left = [x for x in nums if x > pivot]
    mid = [x for x in nums if x == pivot]
    right = [x for x in nums if x < pivot]
    
    L_len = len(left)
    M_len = len(mid)
    
    # 第 k 大的元素在大于基准的部分
    if k <= L_len:
        return findKthLargest_pythonic(left, k)
    # 第 k 大的元素正好等于基准值
    elif k > L_len + M_len:
        # 第 k 大的元素在小于基准的部分，需要更新 k 的值
        return findKthLargest_pythonic(right, k - L_len - M_len)
    else:
        return mid[0]
        
# 测试
arr = [3, 2, 3, 1, 2, 4, 5, 5, 6]
print(findKthLargest_pythonic(arr, 4))  # 输出: 4
```
